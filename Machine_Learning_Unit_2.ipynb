```python
#1. Load the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/datasets/covid-19/main/data/time-series-19-covid-combined.csv")
dataset.head(5)
```

```python
#2. show row and column
dataset.shape
```

```python
#Describe the dataset
dataset.describe()
```

```python
#4. Check the variable(numerical / categorical)
numerical_variable = dataset.columns[dataset.dtypes != "object"]
dataset[numerical_variable].head(5)
```

```python
categorical_variable = dataset.columns[dataset.dtypes == "object"]
dataset[categorical_variable].head(5)
```

```python
#4. Missing value
dataset.isnull().head(5)
```

```python
#specific for numerical value
dataset[numerical_variable].isnull().head(5)
```

```python
#Find out the percentage of missing values
dataset[numerical_variable].isnull().sum()
dataset[numerical_variable].isnull().sum()/len(dataset)
```

```python
#Replace the missing values
dataset.fillna('KIT')
```

```python
#replace specific column
dataset['Confirmed'].replace(0, 1).head(5)
```

```python
dataset.replace(0,"KIT")
```

```python
#Replace more than one value
dataset.replace([0,1], ["KIT", 0]).head(15)
```

```python
#create the row and add it with the original dataset
#Value of variable , instructions about the variable
row_insert = pd.Series(['2022-05-20', 'Cambodia', 'TTY', 2000, 500, 0], index = ['Date', 'Country/Region', 'Province/State', 'Confirmed', 'Recovered', 'Deaths'])
dataset.append(row_insert, ignore_index=True)
```

```python
#call specific row
#datasetname.iloc[] = base on index
dataset.iloc[0]
dataset.iloc[1:5]
```

```python

#loc = to call the specific values,
#index
dataset = dataset.set_index(dataset['Date'])
dataset.loc['2020-01-22']
```

```python
#rename
dataset.rename(columns={'Country/Region': 'Country'}).head(5)
```

```python
#delete the columns
dataset.drop('Province/State', axis=1).head(5) 
dataset.drop(dataset.columns[1], axis=1)
```

```python
#delete row
dataset.drop([1], axis = 0)
```
